```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import statsmodels.api as sm
from numpy.polynomial.polynomial import polyfit
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv("C:/Users/Yogendra/Downloads/delivery_time.csv")
```

```python
data = data.rename({'Delivery Time':'Deliverytime'},axis =1)
data = data.rename({'Sorting Time':'Sortingtime'},axis =1)
```

```python
plt.plot(data.Sortingtime,data.Deliverytime,'bo')
```

```python
data.corr()
```

```python
x = data['Sortingtime']
y = data['Deliverytime']
```

```python
b, m = polyfit(x, y, 1)
```

```python
plt.scatter(x, y)
plt.plot(x, y, '.')
plt.plot(x, b + m * x, '-')
plt.title('Scatter plot Delivery_Time')
plt.xlabel('Sorting_Time')
plt.ylabel('Delivery_Time')
plt.show()
```

```python
#Model Building
model = sm.OLS(y, x).fit()
predictions = model.predict(x)
model.summary()
```

```python
#X log transformation
xlog = np.log(data['Sortingtime'])
```

```python
xlogmodel = sm.OLS(y, xlog).fit()
predictions = xlogmodel.predict(xlog)
xlogmodel.summary()
```

```python
#Y log transformation
ylog = np.log(data['Deliverytime'])
ylogmodel = sm.OLS(ylog, x).fit()
predictions = ylogmodel.predict(x)
model.summary()
```

```python
#X,Y log transformation
xylogmodel = sm.OLS(ylog, xlog).fit()
predictions = xylogmodel.predict(xlog)
model.summary()
```

```python
#X sqrt transformation
sqrtx = np.sqrt(data['Sortingtime'])
```

```python
sqrtxmodel = sm.OLS(y, sqrtx).fit()
predictions = sqrtxmodel.predict(sqrtx)
sqrtxmodel.summary()
```

```python
#Y sqrt transformation
sqrty = np.sqrt(data['Deliverytime'])
```

```python
sqrtymodel = sm.OLS(sqrty, x).fit()
predictions = sqrtymodel.predict(x)
sqrtymodel.summary()
```

```python
# X, Y Root Transformation
sqrtxymodel = sm.OLS(sqrty, sqrtx).fit()
predictions = model.predict(sqrtx)
sqrtxymodel.summary()
```
